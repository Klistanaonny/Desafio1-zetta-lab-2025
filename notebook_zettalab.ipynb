## Popula√ß√£o

Cerca de 84,37% da popula√ß√£o de Melga√ßo (PA) √© composta por pessoas pretas. Esse dado permite compreender como fatores √©tnico-raciais impactam diretamente o acesso √† educa√ß√£o e, consequentemente, o desenvolvimento econ√¥mico. O munic√≠pio de Melga√ßo representa, portanto, um exemplo concreto de como desigualdades hist√≥ricas podem se refletir nos indicadores socioecon√¥micos brasileiros.
A heran√ßa da escravid√£o, que perdurou por s√©culos no pa√≠s, deixou marcas profundas na estrutura social brasileira. Mesmo ap√≥s a aboli√ß√£o da escravid√£o em 1888, o racismo estrutural continuou limitando as oportunidades de pessoas pretas, que muitas vezes foram exclu√≠das do mercado formal de trabalho e privadas do acesso √† educa√ß√£o.
Essa exclus√£o hist√≥rica gerou ciclos de pobreza: fam√≠lias negras, com menor renda e acesso √† informa√ß√£o, enfrentam maiores taxas de natalidade e menores condi√ß√µes de escolariza√ß√£o. Assim, √© poss√≠vel perceber que as desigualdades raciais interferem diretamente na educa√ß√£o e, por consequ√™ncia, no desenvolvimento econ√¥mico de munic√≠pios como Melga√ßo.
Em contextos onde crian√ßas precisam trabalhar para contribuir com a renda familiar, o estudo acaba sendo deixado de lado, o que perpetua a falta de qualifica√ß√£o e
 mant√©m baixos os √≠ndices de renda. Dessa forma, percebe-se que a quest√£o racial est√° profundamente interligada √† desigualdade educacional e econ√¥mica no Brasil.

```python
# Importando as bibliotecas necess√°rias
# matplotlib e seaborn para visualiza√ß√£o de dados, pandas para manipula√ß√£o tabular
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Criando um dicion√°rio com os dados de distribui√ß√£o racial em Melga√ßo (PA)
# As chaves s√£o os nomes das colunas e os valores s√£o listas com os dados correspondentes
# Dados
dados = {
    "Cor/Ra√ßa": ["Pretos", "Pardos", "Brancos", "Outros"],
    "Percentual": [84.37, 10.15, 4.2, 1.28]
}

# Convertendo o dicion√°rio em um DataFrame (estrutura de dados do pandas)
df = pd.DataFrame(dados)


# Definindo o estilo visual do gr√°fico com o tema "whitegrid" e paleta de cores "viridis"
sns.set(style="whitegrid", palette="viridis")

# Criando a figura e definindo o tamanho
plt.figure(figsize=(8, 5))


# Criando o gr√°fico de barras com seaborn
# Eixo X: categorias de cor/ra√ßa | Eixo Y: percentuais da popula√ß√£o
grafico = sns.barplot(
    x="Cor/Ra√ßa",
    y="Percentual",
    data=df
)

# Adicionando os valores percentuais acima de cada barra para facilitar a leitura
for i, valor in enumerate(df["Percentual"]):
    plt.text(i, valor + 1, f"{valor:.2f}%", ha="center", fontsize=11, fontweight="bold")

# Adicionando t√≠tulo e r√≥tulos aos eixos
plt.title("Distribui√ß√£o Racial da Popula√ß√£o de Melga√ßo (PA)", fontsize=14, weight="bold")
plt.ylabel("Percentual (%)")
plt.xlabel("")

# Ajustando o limite do eixo Y para dar espa√ßo aos r√≥tulos
plt.ylim(0, 100)

# Ajustando margens e espa√ßamento do gr√°fico
plt.tight_layout()

# Exibindo o gr√°fico
plt.show()
```

## Raz√£o de depend√™ncia total

De acordo com o Atlas Brasil, a raz√£o de depend√™ncia total ‚Äî indicador que representa o n√∫mero de pessoas dependentes (crian√ßas e idosos) em rela√ß√£o √† popula√ß√£o em idade ativa (15 a 64 anos) ‚Äî apresentou queda no munic√≠pio de Melga√ßo (PA) entre 2000 e 2010, passando de 114,11% para 88,05%.
Nesse mesmo per√≠odo, a propor√ß√£o de idosos diminuiu de 3,01% para 2,61%.
Essa redu√ß√£o indica que h√° menos pessoas chegando √† velhice, o que pode estar relacionado √† baixa expectativa de vida no munic√≠pio. Muitos habitantes precisam come√ßar a trabalhar desde cedo, em atividades de baixa remunera√ß√£o e alta exig√™ncia f√≠sica, como a pesca e o extrativismo.
Essa realidade reflete o baixo acesso √† educa√ß√£o e √† qualifica√ß√£o profissional, o que limita as oportunidades e mant√©m grande parte da popula√ß√£o em condi√ß√µes prec√°rias de trabalho e renda.
Com o passar dos anos, essa sobrecarga e falta de acesso a servi√ßos de sa√∫de adequados reduzem a qualidade de vida, fazendo com que poucos moradores cheguem √† terceira idade. Assim, os dados da raz√£o de depend√™ncia total refor√ßam a conex√£o entre educa√ß√£o, condi√ß√µes socioecon√¥micas e longevidade no munic√≠pio de Melga√ßo.

```python

# Importando as bibliotecas necess√°rias
# pandas para manipula√ß√£o de dados tabulares
# matplotlib e seaborn para visualiza√ß√£o de dados
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Criando o DataFrame com os dados da raz√£o de depend√™ncia total
# Esse indicador mostra a propor√ß√£o de pessoas dependentes (crian√ßas e idosos)
# em rela√ß√£o √† popula√ß√£o em idade ativa (15 a 64 anos)
dados_dependencia = pd.DataFrame({
    "Ano": [2000, 2010],
    "Raz√£o de Depend√™ncia Total (%)": [114.11, 88.05]
})

# Criando o DataFrame com os dados da propor√ß√£o de idosos no munic√≠pio
# Indica a porcentagem de pessoas com 65 anos ou mais em cada ano
dados_idosos = pd.DataFrame({
    "Ano": [2000, 2010],
    "Propor√ß√£o de Idosos (%)": [3.01, 2.61]
})

# Definindo o estilo visual dos gr√°ficos com seaborn
sns.set(style="whitegrid", palette="crest")

# ==============================
# üìâ Gr√°fico 1 ‚Äî Raz√£o de Depend√™ncia Total
# ==============================
plt.figure(figsize=(8,5))

# Criando um gr√°fico de linha para mostrar a varia√ß√£o entre 2000 e 2010
sns.lineplot(
    x="Ano",
    y="Raz√£o de Depend√™ncia Total (%)",
    data=dados_dependencia,
    marker="o",
    linewidth=2.5
)

# T√≠tulo e r√≥tulos dos eixos
plt.title("Queda da Raz√£o de Depend√™ncia Total ‚Äî Melga√ßo (PA)", fontsize=14, weight="bold")
plt.ylabel("Percentual (%)")
plt.xlabel("Ano")

# Definindo os limites do eixo Y para melhorar a visualiza√ß√£o
plt.ylim(80, 120)

# Adicionando os valores percentuais acima dos pontos
for i, v in enumerate(dados_dependencia["Raz√£o de Depend√™ncia Total (%)"]):
    plt.text(dados_dependencia["Ano"][i], v + 1, f"{v:.2f}%", ha='center', fontsize=11, fontweight="bold")

# Ajustando margens e exibindo o gr√°fico
plt.tight_layout()
plt.show()

# ==============================
# üëµ Gr√°fico 2 ‚Äî Propor√ß√£o de Idosos
# ==============================
plt.figure(figsize=(6,5))

# Criando um gr√°fico de barras para comparar a propor√ß√£o de idosos nos dois anos
sns.barplot(
    x="Ano",
    y="Propor√ß√£o de Idosos (%)",
    data=dados_idosos,
    palette="crest"
)

# T√≠tulo e r√≥tulos dos eixos
plt.title("Propor√ß√£o de Idosos ‚Äî Melga√ßo (PA)", fontsize=14, weight="bold")
plt.ylabel("Percentual (%)")
plt.ylim(0, 4)
for i, v in enumerate(dados_idosos["Propor√ß√£o de Idosos (%)"]):
    plt.text(i, v + 0.1, f"{v:.2f}%", ha='center', fontsize=11, fontweight="bold")

# Ajustando margens e exibindo o gr√°fico
plt.tight_layout()
plt.show()
```

O gr√°fico mostra uma queda significativa tanto na raz√£o de depend√™ncia total quanto na propor√ß√£o de idosos, o que refor√ßa a baixa expectativa de vida e a precariedade das condi√ß√µes de trabalho e sa√∫de em Melga√ßo.

## Pir√¢mide et√°ria

Ao observar a pir√¢mide et√°ria de Melga√ßo (PA) dispon√≠vel no Atlas Brasil, nota-se uma base larga que se estreita rapidamente ao longo da estrutura. Isso indica que o munic√≠pio possui alta taxa de natalidade e uma popula√ß√£o predominantemente jovem.
Uma das principais causas desse padr√£o pode estar relacionada √† gravidez na adolesc√™ncia. Muitos jovens carecem de acesso √† informa√ß√£o sobre sexualidade, m√©todos contraceptivos e preven√ß√£o de infec√ß√µes sexualmente transmiss√≠veis.
A falta desse conhecimento b√°sico leva a gesta√ß√µes precoces, e, diante da necessidade de sustento, muitas dessas jovens m√£es e pais abandonam os estudos para trabalhar. Esse cen√°rio reduz a qualifica√ß√£o profissional e perpetua o ciclo de pobreza na regi√£o.
Al√©m disso, parte da popula√ß√£o n√£o tem clareza sobre as consequ√™ncias socioecon√¥micas de uma gravidez precoce, tanto para a fam√≠lia quanto para o munic√≠pio. O aumento do n√∫mero de filhos em contextos de vulnerabilidade tende a sobrecarregar a renda familiar, dificultar o acesso a servi√ßos b√°sicos e impactar negativamente o desenvolvimento econ√¥mico local.
Dessa forma, a pir√¢mide et√°ria de Melga√ßo reflete n√£o apenas um fen√¥meno demogr√°fico, mas tamb√©m as desigualdades educacionais e informacionais que afetam diretamente o futuro de sua popula√ß√£o.

```python
# Importando as bibliotecas necess√°rias
# pandas: manipula√ß√£o e organiza√ß√£o de dados
# matplotlib e seaborn: cria√ß√£o e personaliza√ß√£o dos gr√°ficos
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Criando listas com as faixas et√°rias e os percentuais de homens e mulheres
# Esses valores s√£o simulados para representar o formato da pir√¢mide et√°ria de Melga√ßo (PA)
faixas = ["0‚Äì4", "5‚Äì9", "10‚Äì14", "15‚Äì19", "20‚Äì24", "25‚Äì29", "30‚Äì39", "40‚Äì49", "50‚Äì59", "60+"]
homens = [8, 7.5, 7, 6.5, 6, 5.5, 4, 3, 2, 1]
mulheres = [7.5, 7, 6.8, 6.2, 6, 5.7, 4.2, 3.2, 2.2, 1.2]

# Criando um DataFrame com os dados simulados
# Cada linha representa uma faixa et√°ria e as colunas indicam o percentual de homens e mulheres
df = pd.DataFrame({
    "Faixa Et√°ria": faixas,
    "Homens": homens,
    "Mulheres": mulheres
})

# Definindo o estilo visual do gr√°fico
sns.set(style="whitegrid")

# Criando a figura e ajustando o tamanho
plt.figure(figsize=(8,6))

# Criando as barras horizontais da pir√¢mide et√°ria
# As barras dos homens s√£o negativas para que apare√ßam √† esquerda
plt.barh(df["Faixa Et√°ria"], -df["Homens"], color="#4C72B0", label="Homens")
# As barras das mulheres s√£o positivas (√† direita)
plt.barh(df["Faixa Et√°ria"], df["Mulheres"], color="#DD8452", label="Mulheres")

# Adicionando t√≠tulo e r√≥tulos dos eixos
plt.title("Pir√¢mide Et√°ria de Melga√ßo (PA) ‚Äî Representa√ß√£o", fontsize=14, weight="bold")
plt.xlabel("Percentual da Popula√ß√£o (%)")
plt.ylabel("Faixa Et√°ria")

# Ajustando o eixo X para mostrar valores absolutos sim√©tricos dos dois lados
plt.xlim(-10, 10)
plt.xticks(range(-10, 11, 2), [abs(x) for x in range(-10, 11, 2)])

# Adicionando legenda para diferenciar homens e mulheres
plt.legend(loc="lower right")

# Ajustando margens e espa√ßamento
plt.tight_layout()

# Exibindo o gr√°fico
plt.show()

```

O formato da pir√¢mide ‚Äî com base larga e topo estreito ‚Äî indica que poucos habitantes chegam √† velhice, reflexo da baixa qualidade de vida, da desigualdade e do trabalho precoce.

## Mortalidade infantil

A mortalidade infantil representa o n√∫mero de crian√ßas que morrem antes de completar um ano de idade, em rela√ß√£o ao total de nascidos vivos em determinado per√≠odo.
De acordo com o Atlas Brasil, o √≠ndice de Melga√ßo (PA) apresentou redu√ß√£o significativa entre 2000 e 2010 ‚Äî passando de 45,39 para 22,40 mortes por mil nascidos vivos.
Apesar dessa melhora, o valor ainda est√° acima da meta estabelecida pela ONU, que prev√™ menos de 12 mortes por mil at√© 2030.
Esse indicador continua sendo motivo de preocupa√ß√£o no munic√≠pio e pode ser explicado, em parte, pelas condi√ß√µes socioecon√¥micas e educacionais da popula√ß√£o.
A falta de informa√ß√£o sobre sa√∫de e planejamento familiar faz com que muitas gesta√ß√µes ocorram sem acompanhamento m√©dico adequado, e a aus√™ncia de pr√©-natal aumenta os riscos durante e ap√≥s o parto.
Al√©m disso, grande parte das fam√≠lias n√£o possui renda suficiente para garantir cuidados b√°sicos nos primeiros meses de vida do beb√™, como a realiza√ß√£o do teste do pezinho, essencial para detectar precocemente doen√ßas gen√©ticas e metab√≥licas.
Assim, observa-se que a mortalidade infantil em Melga√ßo est√° fortemente ligada √† desigualdade social, √† desinforma√ß√£o e √† precariedade dos servi√ßos de sa√∫de, perpetuando o ciclo de vulnerabilidade do munic√≠pio.

```python
# Importando as bibliotecas necess√°rias
# pandas: manipula√ß√£o e organiza√ß√£o dos dados
# matplotlib e seaborn: cria√ß√£o e personaliza√ß√£o dos gr√°ficos
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Criando um DataFrame com os dados de mortalidade infantil de Melga√ßo (PA)
# O indicador representa o n√∫mero de crian√ßas que morrem antes de completar um ano de idade
# a cada mil nascidos vivos
dados = pd.DataFrame({
    "Ano": [2000, 2010],
    "Mortalidade Infantil (por mil nascidos vivos)": [45.39, 22.40]
})

# Definindo o estilo visual do gr√°fico
sns.set(style="whitegrid", palette="crest")

# Criando a figura e ajustando o tamanho
plt.figure(figsize=(8,5))

# Criando a linha principal que mostra a evolu√ß√£o da mortalidade infantil entre 2000 e 2010
sns.lineplot(
    data=dados,
    x="Ano",
    y="Mortalidade Infantil (por mil nascidos vivos)",
    marker="o",          # adiciona marcadores nos pontos
    linewidth=2.5,       # espessura da linha
    color="#2E86AB",     # cor personalizada
    label="Melga√ßo (PA)" # legenda da linha principal
)

# Adicionando uma linha de refer√™ncia com a meta estabelecida pela ONU (12 mortes por mil)
# Essa linha serve para comparar o desempenho de Melga√ßo com o objetivo global at√© 2030
plt.axhline(12, color="#E74C3C", linestyle="--", linewidth=2, label="Meta da ONU (2030)")

# Definindo o t√≠tulo e os r√≥tulos dos eixos
plt.title("Redu√ß√£o da Mortalidade Infantil ‚Äî Melga√ßo (PA)", fontsize=14, weight="bold")
plt.ylabel("Mortes por mil nascidos vivos")
plt.xlabel("Ano")

# Definindo os limites do eixo Y para melhor visualiza√ß√£o
plt.ylim(0, 50)

# Adicionando os valores num√©ricos sobre os pontos do gr√°fico
for i, v in enumerate(dados["Mortalidade Infantil (por mil nascidos vivos)"]):
    plt.text(dados["Ano"][i], v + 1, f"{v:.2f}", ha='center', fontsize=11, fontweight="bold")

# Exibindo a legenda, ajustando o layout e mostrando o gr√°fico
plt.legend()
plt.tight_layout()
plt.show()

```

Apesar da melhora, o valor ainda √© quase o dobro da meta da ONU para 2030 (12 mortes por mil). 
Essa diferen√ßa reflete as dificuldades de acesso √† sa√∫de, educa√ß√£o e renda ‚Äî fatores que perpetuam a vulnerabilidade social no munic√≠pio.

A elevada porcentagem de crian√ßas matriculadas entre 5 e 6 anos pode estar associada ao fato de que muitos pais utilizam a escola como um espa√ßo seguro enquanto trabalham, o que permite maior disponibilidade de tempo para atividades laborais. Os dados mostram que, embora haja pequenas redu√ß√µes nos √≠ndices, o problema da perman√™ncia escolar continua estrutural.

Embora esse n√∫mero possa parecer razo√°vel diante do contexto socioecon√¥mico do munic√≠pio, ele revela que mais de um ter√ßo dos habitantes ainda vive sem um recurso essencial √† sa√∫de e √† dignidade humana.

De acordo com um document√°rio produzido pela BBC News, √© poss√≠vel observar fisicamente que nenhuma resid√™ncia do munic√≠pio possui tratamento de esgoto, o que evidencia de forma concreta a gravidade da situa√ß√£o sanit√°ria em Melga√ßo.

A compara√ß√£o evidencia como o acesso √† infraestrutura b√°sica √© um marcador direto da desigualdade social e econ√¥mica no pa√≠s.
Entende-se, portanto, que vulnerabilidade social n√£o se limita √† aus√™ncia de renda, mas √† exposi√ß√£o constante a condi√ß√µes prec√°rias de moradia, sa√∫de e educa√ß√£o.
Popula√ß√µes vulner√°veis, como a de Melga√ßo, est√£o mais sujeitas a problemas de sa√∫de, evas√£o escolar e exclus√£o social, o que perpetua um ciclo de pobreza dif√≠cil de romper. Dessa forma, a an√°lise do acesso √† √°gua encanada, ao esgoto e √† habita√ß√£o adequada refor√ßa que as desigualdades estruturais do Brasil v√£o al√©m da economia, alcan√ßando dimens√µes b√°sicas da sobreviv√™ncia humana e comprometendo o desenvolvimento social e econ√¥mico a longo prazo.

```python
# Importando as bibliotecas necess√°rias
# pandas: para manipula√ß√£o de dados em formato de tabela
# matplotlib e seaborn: para cria√ß√£o e customiza√ß√£o de gr√°ficos
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Criando um DataFrame com os dados sobre acesso √† √°gua e esgoto em duas cidades
# Os dados evidenciam a diferen√ßa de infraestrutura entre Melga√ßo (PA) e S√£o Caetano do Sul (SP)
dados = pd.DataFrame({
    "Cidade": ["Melga√ßo (PA)", "Melga√ßo (PA)", "S√£o Caetano do Sul (SP)", "S√£o Caetano do Sul (SP)"],
    "Indicador": ["√Ågua encanada", "Esgoto tratado", "√Ågua encanada", "Esgoto tratado"],
    "Percentual (%)": [67.66, 0, 100, 100]
})

# Definindo o estilo visual do gr√°fico
sns.set(style="whitegrid", palette="crest")

# Criando a figura e ajustando o tamanho
plt.figure(figsize=(9,5))

# Criando um gr√°fico de barras agrupadas para comparar os dois indicadores em cada cidade
sns.barplot(
    data=dados,
    x="Cidade",
    y="Percentual (%)",
    hue="Indicador",                # separa as barras por tipo de infraestrutura
    palette=["#2E86AB", "#85C1E9"]  # tons diferentes para √°gua e esgoto
)

# Adicionando os valores percentuais acima de cada barra
# O c√°lculo da posi√ß√£o (i//2 e ¬±0.18) centraliza os textos sobre as barras correspondentes
for i, v in enumerate(dados["Percentual (%)"]):
    plt.text(i//2 + (0.18 if i % 2 == 0 else -0.18), v + 2, f"{v:.0f}%", 
             ha='center', fontsize=11, fontweight="bold")

# Adicionando t√≠tulo e r√≥tulos dos eixos
plt.title("Acesso √† Infraestrutura B√°sica ‚Äî Melga√ßo (PA) vs S√£o Caetano do Sul (SP)", 
          fontsize=14, weight="bold")
plt.ylabel("Percentual da Popula√ß√£o (%)")

# Definindo limite do eixo Y para deixar espa√ßo para os r√≥tulos
plt.ylim(0, 110)

# Exibindo a legenda e organizando o layout final
plt.legend(title="Indicador")
plt.tight_layout()

# Exibindo o gr√°fico
plt.show()

```

Essa diferen√ßa mostra como o saneamento √© um marcador direto da desigualdade social e econ√¥mica.
